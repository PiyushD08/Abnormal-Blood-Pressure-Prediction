```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler().fit_transform(x)
scale
```

### Sampling the data into train and test split.

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(scale, y, test_size= 0.25, random_state=229)  
```

```python
x_train.shape
```

```python
x_test.shape
```

# Logistic Regression

```python
#Fitting Logistic Regression to the training set  
from sklearn.linear_model import LogisticRegression  
clf_lr= LogisticRegression(max_iter=100000).fit(x_train,y_train)  
```

```python
#Predicting the test set result  
y_pred= clf_lr.predict(x_test)
```

```python
#Creating the Confusion matrix
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(cf_matrix, annot=True, fmt="d")

plt.show()
```

```python
# Checking Accuracy, Precision, Recall and F1 score
from sklearn.metrics import accuracy_score, precision_score, recall_score,f1_score 
from sklearn.metrics import roc_auc_score, roc_curve

acc_lr = accuracy_score(y_test, y_pred)*100
prsn_lr = precision_score(y_test,y_pred)*100
r_lr= recall_score(y_test, y_pred)*100
f1_lr = f1_score(y_test, y_pred)*100
Auroc = roc_auc_score(y_test,y_pred)*100

print("Accuracy (Logistic Regression):", acc_lr)
print('Presision Score (Logistic Regression):', prsn_lr)
print("Recall Score (Logistic Regression):", r_lr)
print("F1 score (Logistic Regression):", f1_lr)
print('The Roc_auc score (Logistic Regression):', Auroc)
```

```python
pred_prob = clf_lr.predict_proba(x_test)
pred_prob
```

```python
fpr, tpr, thresh  = roc_curve(y_test, pred_prob[:,1])
fpr
```

```python
plt.plot(fpr, tpr, label="Logistic Regression")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver Operating Characteristic")
plt.legend(loc="lower right")
plt.text(x = 0.5, y = 0.5, s = "AUC: 0.7254")
plt.show()
```
